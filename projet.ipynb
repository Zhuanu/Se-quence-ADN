```python
#determine si une paire est concordante
def paire_concordante(a,b):
   return ((a, b) == ('A', 'T') or (a, b) == ('T', 'A') or (a,b) == ('C', 'G') or (a, b) == ('G', 'C'))
```

```python
#Test paire_concordante
assert(paire_concordante('T', 'A'))
assert(paire_concordante('A', 'T'))
assert(not paire_concordante('A', 'C'))
```

```python
#determine le coût d'une opération op entre a et b
def cout(op, a, b):
   if(op == 'del' or op =='ins'):
      return 2
   if (a == b):
      return 0   
   elif (paire_concordante(a, b)):
      return 3
   return 4  
```

```python
def cout2(a, b):
   if(a != '-' and b != '-'): op='sub'
   elif(a == '-'): op = 'ins'
   else: op = 'del'
   return cout(op, a, b)
```

```python
#DIST_NAIF_REC non optimisé avec copie de liste à chaque recursion
def DIST_NAIF_REC(x, y, i, j, c, dist):

   if (i == len(x)-1 and j == len(y)-1):
      if (c < dist):
         dist = c

   else:
      if(i < len(x)-1 and j < len(y)-1):
         dist = DIST_NAIF_REC(x, y, i+1 ,j+1 ,c + cout('sub', x[i], y[j]), dist)
      if(i < len(x)-1):
         dist =  DIST_NAIF_REC(x, y, i+1, j, c + cout('del', '', ''), dist)
      if(j < len(y)-1):
         dist = DIST_NAIF_REC(x, y, i, j+1, c + cout('ins', '', ''), dist)
         
   return dist      
```

```python
#DIST_NAIF_REC optimisé avec des sub-list 
def DIST_NAIF_REC2(x, y, i, j, c, dist):

   if (i == len(x)-1 and j == len(y)-1):
      if (c < dist):
         dist = c

   else:
      if (i < len(x)-1 and j < len(y)-1):
         dist = DIST_NAIF_REC2(x[i+1:], y[j+1:], 0, 0, c + cout('sub', x[i], y[j]), dist)
      if (i < len(x)-1):
         dist =  DIST_NAIF_REC2(x[i+1:], y, 0, 0, c + cout('del', '', ''), dist)
      if (j < len(y)-1):
         dist = DIST_NAIF_REC2(x, y[j+1:], 0, 0, c + cout('ins', '', ''), dist) 
         
   return dist
```

```python
def DIST_NAIF(x,y):
   return DIST_NAIF_REC(x, y, 0, 0, 0, 99999999)
```

```python
def DIST_1(x, y):
   global D
   D=[[-1 for j in range(len(y)+1)] for i in range(len(x)+1)] 
   
   for i in range(len(x)+1):
      for j in range(len(y)+1):
         if (i == 0 and j == 0): D[i][j] = 0;
         elif (i == 0): D[i][j] = 2 * j;
         elif (j == 0): D[i][j] = 2 * i;
         else: 
            m1 = D[i-1][j-1] + cout('sub', x[i-1], y[j-1])
            m2 = D[i-1][j] + cout('del','','')
            m3 = D[i][j-1] + cout('ins','','')
            D[i][j] = min(m1, m2, m3)
   return D[i][j]      
         
```

```python
def SOL_1(x,y,D):
   u = ""
   v = ""
   i = len(x)
   j = len(y)
   while (i != 0 or j != 0):
         if (D[i][j] == D[i][j-1] + cout('ins', '', '')):
            u = '-' + u
            v = y[j-1] + v
            j -= 1
         elif(D[i][j] == D[i-1][j] + cout('del', '', '')):
            u = x[i-1] + u
            v = '-' + v
            i -= 1
         elif(D[i][j] == D[i-1][j-1] + cout('sub', x[i-1], y[j-1])):
            u = x[i-1] + u
            v = y[j-1] + v
            i -= 1
            j -= 1   
   return (u, v)           
```

```python
def PROG_DYN(x, y):
   global D
   d = DIST_1(x, y)
   u,v = SOL_1(x, y, D)
   return (d, u, v)
   
```

```python
def mot_gaps(k):
   resultat = ""
   for i in range(k):
      resultat += "-"
   return resultat   
```

```python
def align_lettre_mot(x,y):
   meilleur_indice = 0
   meilleur_coût = 4
   for i in range(len(y)):
      if (cout2(x, y[i]) < meilleur_coût):
         meilleur_indice = i
         meilleur_coût = cout2(x, y[i])
   return (mot_gaps(meilleur_indice) + x + mot_gaps(len(y) - meilleur_indice-1), y)
```

```python
import math
def coupure(x, y):
   global D
   global K
   K=[[-1 for j in range(len(y)+1)] for i in range(2)] 
   D=[[-1 for j in range(len(y)+1)] for i in range(2)] 
   
   n = len(x)  
   m = len(y)
   i_etoile = math.floor(n/2)
   #Initialisation des valeurs de k
   for z in range(len(y)+1):
      K[0][z] = z 

   for i in range(len(x)+1):
      for j in range(len(y)+1):
         if (i == 0 and j == 0): D[1][j] = 0;
         elif(i == 0): D[1][j] = 2 * j;
         elif(j == 0): 
            D[1][j] = 2 * i
            K[1][j] = K[0][j]
         else: 
            m1 = D[0][j-1] + cout('sub', x[i-1], y[j-1])
            m2 = D[0][j] + cout('del', '', '')
            m3 = D[1][j-1] + cout('ins', '', '')
            D[1][j] = min(m1, m2, m3)
            if i > i_etoile :
               if D[1][j] == m1:
                  K[1][j] = K[0][j-1]
               elif D[1][j] == m3 :
                  K[1][j] = K[1][j-1]
               else:
                  K[1][j] = K[0][j]            
      D.pop(0)  
      D.insert(1, [-1 for j in range(len(y)+1)])
      if i > i_etoile :
         K.pop(0)
         K.insert(1, [-1 for j in range(len(y)+1)])
  
   return K[0][len(y)] 
```

```python
import math
def SOL_2 (x, y) :
   n = len(x) 
   m = len(y)
   if (m == 0):
      return (x, mot_gaps(n))
   if (n == 1):
      return align_lettre_mot(x, y)
   j = coupure(x, y)
   i = math.floor(n/2)
   u,v = SOL_2(x[0:i], y[0:j])
   x,y = SOL_2(x[i:], y[j:])
   return (u + x, v + y)
```
